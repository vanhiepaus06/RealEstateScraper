```python


# Sydney
# 2000, 2006, 2007, 2008, 2009, 2010, 2011, 2015, 2016, 2017, 2018, 2019, 
# 2020, 2021, 2022, 2023, 2024, 2025,  2026, 2027, 2028, 2029, 2030,2031, 
# 2032, 2033, 2034, 2035, 2036, 2037, 2038, 2039, 2040, 2041, 2042, 2043, 
# 2044, 2045, 2046, 2047, 2048, 2049, 2050, 2052, 2060, 2061, 2062, 2063, 
# 2064, 2065, 2066, 2067, 2068, 2069, 2070, 2071, 2072, 2073, 2074, 2075, 
# 2076, 2077, 2079, 2080, 2081, 2082, 2083, 2084, 2085, 2086, 2087, 2088, 
# 2089, 2090, 2092, 2093, 2094, 2095, 2096, 2097, 2099, 2100, 2101, 2102, 
# 2103, 2104, 2105, 2106, 2107, 2108, 2109, 2110, 2111, 2112, 2113, 2114, 
# 2115, 2116, 2117, 2118, 2119, 2120, 2121, 2122, 2123, 2125, 2126, 2127, 
# 2128, 2129, 2130, 2131, 2132, 2133, 2134, 2135, 2136, 2137, 2138, 2139, 
# 2140, 2141, 2142, 2143, 2144, 2145, 2146, 2147, 2148, 2150, 2151, 2152, 
# 2153, 2154, 2155, 2156, 2157, 2158, 2159, 2160, 2161, 2162, 2163, 2164, 
# 2165, 2166, 2167, 2168, 2170, 2171, 2172, 2173, 2174, 2175, 2176, 2177, 
# 2178, 2179, 2190, 2191, 2192, 2193, 2194, 2195, 2196, 2197, 2198, 2199, 
# 2200, 2203, 2204, 2205, 2206, 2207, 2208, 2209, 2210, 2211, 2212, 2213, 
# 2214, 2216, 2217, 2218, 2219, 2220, 2221, 2222, 2223, 2224, 2225, 2226, 
# 2227, 2228, 2229, 2230, 2231, 2232, 2233, 2234, 2555, 2556, 2557, 2558, 
# 2559, 2560, 2563, 2564, 2565, 2566, 2567, 2568, 2569, 2570, 2571, 2572, 
# 2573, 2574, 2745, 2747, 2748, 2749, 2750, 2752, 2753, 2754, 2755, 2756, 
# 2757, 2758, 2759, 2760, 2761, 2762, 2763, 2765, 2766, 2767, 2768, 2769, 
# 2770, 2773, 2774, 2775, 2776, 2777, 2778, 2779, 2780, 2782, 2783, 2784, 
# 2785, 2786, 2787, 2790


# Central Coast
# 2083, 2250, 2251, 2256, 2257, 2258, 2259, 2260, 2261, 2262, 2263, 2775


# Newcastle and Lake Macquarie
# 2259, 2264, 2265, 2267, 2278, 2280, 2281, 2282, 2283, 2284, 2304, 2305, 
# 2306, 2307, 2297, 2298, 2299, 2300, 2302, 2303, 
# 2285, 2286, 2287, 2289, 2290, 2291, 2292, 2293, 2294, 2295, 2308, 2318, 2322, 2296, 2323
# 2500, 2502, 2505, 2506, 2508, 2515, 2516, 2517, 2518, 2519, 2522, 2525, 2526, 2527, 2528, 2529, 2530, 2533, 2534, 2535, 2560, 2577,

# 2469, 2470, 2471, 2472, 2473, 2474, 2475, 2476, 2477, 2478, 2479, 2480, 2481, 2482, 2483, 2484, 2485, 2486, 2487, 2488, 2489, 2490




# List of postcodes
postcodes = [
    2000, 2006, 2007, 2008, 2009, 2010, 2011, 2015, 2016, 2017, 2018, 2019, 
2020, 2021, 2022, 2023, 2024, 2025,  2026, 2027, 2028, 2029, 2030,2031, 
2032, 2033, 2034, 2035, 2036, 2037, 2038, 2039, 2040, 2041, 2042, 2043, 
2044, 2045, 2046, 2047, 2048, 2049, 2050, 2052, 2060, 2061, 2062, 2063, 
2064, 2065, 2066, 2067, 2068, 2069, 2070, 2071, 2072, 2073, 2074, 2075, 
2076, 2077, 2079, 2080, 2081, 2082, 2083, 2084, 2085, 2086, 2087, 2088, 
2089, 2090, 2092, 2093, 2094, 2095, 2096, 2097, 2099, 2100, 2101, 2102, 
2103, 2104, 2105, 2106, 2107, 2108, 2109, 2110, 2111, 2112, 2113, 2114, 
2115, 2116, 2117, 2118, 2119, 2120, 2121, 2122, 2123, 2125, 2126, 2127, 
2128, 2129, 2130, 2131, 2132, 2133, 2134, 2135, 2136, 2137, 2138, 2139, 
2140, 2141, 2142, 2143, 2144, 2145, 2146, 2147, 2148, 2150, 2151, 2152, 
2153, 2154, 2155, 2156, 2157, 2158, 2159, 2160, 2161, 2162, 2163, 2164, 
2165, 2166, 2167, 2168, 2170, 2171, 2172, 2173, 2174, 2175, 2176, 2177, 
2178, 2179, 2190, 2191, 2192, 2193, 2194, 2195, 2196, 2197, 2198, 2199, 
2200, 2203, 2204, 2205, 2206, 2207, 2208, 2209, 2210, 2211, 2212, 2213, 
2214, 2216, 2217, 2218, 2219, 2220, 2221, 2222, 2223, 2224, 2225, 2226, 
2227, 2228, 2229, 2230, 2231, 2232, 2233, 2234, 2555, 2556, 2557, 2558, 
2559, 2560, 2563, 2564, 2565, 2566, 2567, 2568, 2569, 2570, 2571, 2572, 
2573, 2574, 2745, 2747, 2748, 2749, 2750, 2752, 2753, 2754, 2755, 2756, 
2757, 2758, 2759, 2760, 2761, 2762, 2763, 2765, 2766, 2767, 2768, 2769, 
2770, 2773, 2774, 2775, 2776, 2777, 2778, 2779, 2780, 2782, 2783, 2784, 
2785, 2786, 2787, 2790, 2083, 2250, 2251, 2256, 2257, 2258, 2259, 2260, 2261, 2262, 2263, 2775,
2259, 2264, 2265, 2267, 2278, 2280, 2281, 2282, 2283, 2284, 2304, 2305, 
2306, 2307, 2297, 2298, 2299, 2300, 2302, 2303, 
2285, 2286, 2287, 2289, 2290, 2291, 2292, 2293, 2294, 2295, 2308, 2318, 2322, 2296, 2323
2500, 2502, 2505, 2506, 2508, 2515, 2516, 2517, 2518, 2519, 2522, 2525, 2526, 2527, 2528, 2529, 2530, 2533, 2534, 2535, 2560, 2577,

2469, 2470, 2471, 2472, 2473, 2474, 2475, 2476, 2477, 2478, 2479, 2480, 2481, 2482, 2483, 2484, 2485, 2486, 2487, 2488, 2489, 2490

    

    
]

# Base URL template
base_url = "https://www.domain.com.au/sold-listings/?ptype=house&postcode={}&excludepricewithheld=1"

# Selenium WebDriver initialization
driver = webdriver.Chrome()

def scrape_data(url, postcode):
    # Lists to store data
    sold_dates = []
    house_prices = []
    address_line_1 = []
    address_line_2 = []
    num_beds = []
    num_baths = []
    num_parkings = []
    square_m = []
    
    # Open the URL
    driver.get(url)
    driver.maximize_window()
    time.sleep(5)

    # Set the number of pages to load and the delay between each load
    pages = 50
    page_delay = 4
    page_sources = []

    for i in range(pages):
        try:
            # Wait until the 'Next Page' button is clickable
            if i == 0:
                # First page, only one button expected
                next_button = WebDriverWait(driver, page_delay).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[data-testid="paginator-navigation-button"]'))
                )
            else:
                # Subsequent pages, look for the second button
                next_buttons = WebDriverWait(driver, page_delay).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[data-testid="paginator-navigation-button"]'))
                )
                if len(next_buttons) < 2:
                    print(f"Less than two buttons found on page {i+1}. Exiting.")
                    break
                next_button = next_buttons[1]

            # Get the page source and add it to the list
            page_sources.append(driver.page_source)
            
            # Scroll to the button to ensure it's clickable and not obscured
            driver.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(1)  # Add a small delay to ensure the element is in view
            
            # Click the 'Next Page' button using JavaScript to avoid interception
            driver.execute_script("arguments[0].click();", next_button)

            # Wait for the new page to load
            time.sleep(page_delay)
            print(f'Page {i+1} of {pages} loaded')
            
        except Exception as e:
            print(f"Error on page {i+1}: {e}. Exiting.")
            break

    # Parse with BeautifulSoup
    for page_source in page_sources:
        soup = BeautifulSoup(page_source, "html.parser")

        # Find all comment containers
        containers = soup.find("div", attrs={"class":"css-1mf5g4s"})

        # Extract data
        for container in containers.find_all("li", class_="css-1qp9106"):
            
            house_sold_date = container.find("span", class_="css-1nj9ymt")
            if house_sold_date:
                sold_dates.append(house_sold_date.text.strip())
            else:
                sold_dates.append("N/A")

            house_price = container.find('p', {'data-testid': 'listing-card-price'})
            if house_price:
                house_prices.append(house_price.text.strip())
            else:
                house_prices.append("N/A")

            address1 = container.find('span', {'data-testid': 'address-line1'})
            if address1:
                address_line_1.append(address1.text.strip())
            else:
                address_line_1.append("N/A")


            address2 = container.find('span', {'data-testid': 'address-line2'})
            if address2:
                address_line_2.append(address2.text.strip())
            else:
                address_line_2.append("N/A")

            # Extract property features
            features = container.find_all('span', {'data-testid': 'property-features-feature'})
            
            if len(features) > 0:
                num_beds.append(features[0].text.strip() if len(features) > 0 else "N/A")
                num_baths.append(features[1].text.strip() if len(features) > 1 else "N/A")
                num_parkings.append(features[2].text.strip() if len(features) > 2 else "N/A")
                square_m.append(features[3].text.strip() if len(features) > 3 else "N/A")
            else:
                num_beds.append("N/A")
                num_baths.append("N/A")
                num_parkings.append("N/A")
                square_m.append("N/A")

    # Create a DataFrame
    df = pd.DataFrame({
        "Sold Date": sold_dates, 
        "Price": house_prices, 
        "Address Line 1": address_line_1,  
        "Address Line 2": address_line_2,
        "Number of Bedrooms": num_beds, 
        "Number of Bathrooms": num_baths, 
        "Number of Parkings": num_parkings, 
        "Square Metres": square_m
    })

    # Save DataFrame to CSV
    df.to_csv(f"real_estate_data_{postcode}.csv", index=False, encoding='utf-8')
    print(f"Data for postcode {postcode} saved to real_estate_data_{postcode}.csv")

# Iterate over postcodes and scrape data
for postcode in postcodes:
    url = base_url.format(postcode)
    scrape_data(url, postcode)

# Close the browser
driver.quit()
```

```python
#Merge all the data files
# Directory where the CSV files are stored
csv_directory = '.'

# List to hold DataFrames
dfs = []

# Loop through the files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv') and filename.startswith('real_estate_data_'):
        # Read each CSV file
        file_path = os.path.join(csv_directory, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate all DataFrames
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('full_real_estate_data_uncleaned.csv', index=False, encoding='utf-8')

print("All CSV files have been merged into 'full_real_estate_data_uncleaned.csv'")
```
